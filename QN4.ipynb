To implement Playfair Cipher
 
* Plaintext will be provided 
* Obtain Ciphertext and Plaintext back.

```python
def create_table(key):
  table = []
  k = []
  for i in key:
    if i not in k:
      if i =="J":
        if 'I' not in k:
          k.append("I")
      else:
        k.append(i)
  for i in range(65,91):
    if chr(i) not in k:
      k.append(i)
  
  for i in range(5):
    table.append(k[:5])
    k = k[5:]
  return table
```

```python
def group2(pt):
  
  groups = []
  i = 0
  while(i < len(pt)-1):
    if(pt[i] != pt[i+1]):
      groups.append(pt[i] + pt[i+1])
      i += 2
    else:
      if pt[i] == 'X':
        t = 'Y'
        groups.append(pt[i]+t)
        i += 1
      else:
        t = 'X'
        groups.append(pt[i]+t)
        i += 1
    if(i == len(pt)-1):
      if(pt[i] == 'X'):
        t = 'Y'
      else:
        t = 'X'
      groups.append(pt[i]+t)
  return groups
```

```python
def search_index(table,e):
  if e == 'J':
    e = 'I'
  else:
    for i in range(5):
      for j in range(5):
        if(table[i][j] == e):
          return i,j
```

```python
pt = "paasyy"
k = ""

group = group2(pt)
table = create_table(k)

for g in group:
  
  c1,c2 = g[0],g[1]
  
  i1,j1 = search_index(table,c1)
  i2,j2 = search_index(table,c2)
  if(i1 == i2):
    ans += table[i1][(j1+1)%5]+table[i2][(j2+1)%5]
  elif(j1 == j2):
    ans += table[(i1+1)%5][j1]+table[(i2+1)%5][j2]
  else:
    ans += table[i1][j2] + table[i2][j1]

ct = ans

print(pt)
```
