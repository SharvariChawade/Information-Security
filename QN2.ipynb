To implement Vignere Cipher (26 Caesar Ciphers)
 
* Plaintext will be provided 
* Obtain Ciphertext and Plaintext back

```python
pt = "aandu gondu"
k = "shari"
```

```python
pt = pt.replace(" ","").upper()
k = k.replace(" ","").upper()
```

```python
def generate_key(k,pt):
  key=''
  j = 0
  for i in range(len(pt)):
    if(j==len(k)):
      j=0
    key += k[j]
    j += 1
  return key

key = generate_key(k,pt)
print(key)
```

```python
def encryption(pt,key):
  ct = ''
  for i in range(len(pt)):
    x = (ord(pt[i])+ord(key[i]))%26
    x += ord('A')
    ct += chr(x)
  # print(ct)
  return ct
```

```python
ct = encryption(pt,key)
print(ct)
```

```python
def decryption(ct,key):
  pt1 = ''
  for i in range(len(ct)):
    x = (ord(ct[i]) - ord(key[i]))%26
    x += ord('A')
    pt1 += chr(x)
  return pt1
```

```python
decryption(ct,key)
```
